```python
import requests
response = requests.get('https://fakestoreapi.com/products')
print(response.json())
```

```python
import pandas as pd

data = response.json()

df = pd.DataFrame(data)

df.head()
```

```python
df.info()
```

```python
df.drop(['image'], axis=1, inplace=True)
```

```python
df.info()
```

```python
df
```

```python
df['rating'].head()
```

```python
import ast

# Convert string to dictionary if needed
df['rating'] = df['rating'].apply(lambda x: ast.literal_eval(x) if isinstance(x, str) else x)

# Separate the columns
df['rate'] = df['rating'].apply(lambda x: x['rate'] if isinstance(x, dict) else None)
df['count'] = df['rating'].apply(lambda x: x['count'] if isinstance(x, dict) else None)
```

```python
df
```

```python
df.drop(['rating'], axis=1, inplace=True)
```

```python
df
```

```python
df['count'] = df['count'].astype(int)
df['rate'] = df['rate'].astype(float)
```

```python
df.info()
```

```python
df.shape
```

```python
df.drop_duplicates(inplace=True)
```

```python
df.isnull().sum()
```
